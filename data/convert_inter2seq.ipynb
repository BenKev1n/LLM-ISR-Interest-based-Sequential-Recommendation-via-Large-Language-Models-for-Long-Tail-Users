```python
import os
import pickle
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
dataset = "LastFM"
```

```python
data = {}
# 读取数据
df = pd.read_csv(f"./{dataset}/handled/inter.txt", header=None, sep = ' ', names = ['user', 'item'])      
```

```python
df2user = df.groupby('user')['item'].apply(lambda x: ' '.join(map(str, x)))
```

```python
# 将结果存储到文本文件中
with open(f"./{dataset}/handled/inter2seq.txt", 'w') as file:
    for index, value in df2user.items():
        file.write(f'{index} {value}\n')
```
