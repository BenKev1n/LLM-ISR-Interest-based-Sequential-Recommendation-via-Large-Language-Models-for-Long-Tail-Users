```python
import os
import pickle
import numpy as np
from tqdm import tqdm
```

```python
dataset = "LastFM"
head_ratio = 0.2
item_emb = pickle.load(open(os.path.join(dataset+"/handled/", "itm_emb_np.pkl"), "rb"))
```

```python
item_int_len = [0] * len(item_emb)
with open(os.path.join(dataset+"/handled/inter.txt"), 'r') as f:
    for line in tqdm(f.readlines()):
        parts = line.strip().split()
        itm_id = int(parts[1])
        item_int_len[itm_id-1] += 1
len(item_int_len)
```

```python
for i in range(10):
    print(item_int_len[i])
```

```python
item_int_len_str = [str(i) + '\n' for i in item_int_len]
len(item_int_len_str)
```

```python
for i in range(10):
    print(item_int_len_str[i])
```

```python
with open(os.path.join(dataset+"/handled/", "itm_inter_freq.txt"), 'w') as f:
    f.writelines(item_int_len_str)
```

```python
itm_inter_freq = {}
index = 0
with open(os.path.join(dataset+"/handled/itm_inter_freq.txt"), 'r') as f:
    for line in tqdm(f.readlines()):
        parts = line.strip().split()
        item_id = int(index)
        item_freq = int(parts[0])
        itm_inter_freq[item_id] = item_freq
        index+=1
len(itm_inter_freq)
```

```python
for i in range(10):
    print(itm_inter_freq[i])
```

```python
sorted_itm_freq = sorted(itm_inter_freq.items(), key=lambda x: x[1], reverse=True)
num_items = len(sorted_itm_freq)
head_item_num = int(num_items * head_ratio)
print(f"Total items: {num_items}, Head users: {head_item_num}, Tail users: {num_items - head_item_num}")
head_items = [i[0] for i in sorted_itm_freq[:head_item_num]]
tail_items = [i[0] for i in sorted_itm_freq[head_item_num:]]
```

```python
head_items_str = [str(i) + '\n' for i in head_items]
tail_items_str = [str(i) + '\n' for i in tail_items]
with open(os.path.join(dataset+"/handled/", "head_item_id.txt"), 'w') as f:
    f.writelines(head_items_str)
with open(os.path.join(dataset+"/handled/", "tail_item_id.txt"), 'w') as f:
    f.writelines(tail_items_str)
```

```python
head_item_emb = {}
for i in range(len(head_items)):
    head_item_emb[i] = item_emb[head_items[i]-1]
print(f"Head item embedding size: {len(head_item_emb)}")
head_item_emb = np.vstack(head_item_emb.values())
print(f"Head user embedding array shape: {head_item_emb.shape}")
pickle.dump(head_item_emb, open(os.path.join(dataset+"/handled/head_item_emb_np.pkl"), "wb"))
```

```python
tail_item_emb = {}
for i in range(len(tail_items)):
    tail_item_emb[i] = item_emb[tail_items[i]-1]
print(f"Tail item embedding size: {len(tail_item_emb)}")
tail_item_emb = np.vstack(tail_item_emb.values())
print(f"Tail user embedding array shape: {tail_item_emb.shape}")
pickle.dump(tail_item_emb, open(os.path.join(dataset+"/handled/tail_item_emb_np.pkl"), "wb"))
```
