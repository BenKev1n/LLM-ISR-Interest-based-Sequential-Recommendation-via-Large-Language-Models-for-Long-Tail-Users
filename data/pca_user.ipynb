```python
import pickle
import os
from sklearn.decomposition import PCA
```

```python
dataset = "LastFM"
```

```python
llm_usr_emb = pickle.load(open(os.path.join(dataset+"/handled/", "usr_emb_np.pkl"), "rb"))
```

```python
llm_usr_emb.shape
```

```python
pca = PCA(n_components=64)
pca_usr_emb = pca.fit_transform(llm_usr_emb)
```

```python
with open(os.path.join(dataset+"/handled/", "pca64_usr_emb_np.pkl"), "wb") as f:
    pickle.dump(pca_usr_emb, f)
```
